```python
#Task 1

class Student:
    id = 0
    def __init__(self, name, dept, age, cg):
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cg
        Student.id +=1
        self.id = Student.id
    def showDetails(self):
        print(f"""ID: {self.id}
Name: {self.name}
Department: {self.dept}
Age: {self.age}
CGPA: {self.cgpa}""")
    @classmethod
    def from_String(cls,string):
        st =string.split('-')
        name = st[0]
        department = st[1]
        age = st[2]
        cgpa = st[3]       
        return Student(name,department,age,cgpa)


s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails() 


#Subtask 5
print("Class variables are created when the program starts and destroyed when the program stops. On the other hand, An instance variable is a variable whose value is specified to the Instance and shared among different instances.")
print("A class method provides functionality to a class itself, while an instance method provides functionality to one instance of a class.")
```

```python
#Task 2
class Assassin:
    num_of_assassin = 0
    def __init__(self, name, percent):
        self.name = name
        self.percent = percent
        Assassin.num_of_assassin += 1
    def printDetails(self):
        print(f"""Name: {self.name}
Success rate: {self.percent}%
Total number of Assassin: {Assassin.num_of_assassin}""")
    @classmethod
    def failureRate(cls, name, frate):
        return Assassin(name, 100 - frate)
    @classmethod
    def failurePercentage(cls, name, frate):

        return Assassin(name, 100 - int(frate[:-1]))

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#Task 7
class Cat:
    Number_of_cats = 0
    def __init__(self, clr, pos):
        self.clr = clr
        self.pos = pos
        Cat.Number_of_cats += 1
    def printCat(self):
        print(f"{self.clr} cat is {self.pos}")
    def changeColor(self, clr):
        self.clr = clr
    @classmethod
    def no_parameter(cls):
        color = "White"
        position = "Sitting"
        return Cat(color, position)
    @classmethod
    def first_parameter(cls, color):
        position = "Sitting"
        return Cat(color, position)
    @classmethod
    def second_parameter(cls, position):
        color = "Grey"
        return Cat(color, position)
    
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        self.radius = radius
        self.height = height
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")
    @staticmethod
    def area(radius, height):
        area =  (2*math.pi*radius**2) + (2*math.pi*radius*height)
        print(f"Area: {area}")
    @staticmethod
    def volume(radius, height):
        volume =  (math.pi*radius**2*height)
        print(f"Volume: {volume}")
    @classmethod
    def swap(cls, height, radius):
        return Cylinder(radius, height)
    @classmethod
    def changeFormat(cls, string):
        r,h = string.split("-")
        return Cylinder(float(r), float(h))

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 10
class SultansDine:
    total = 0
    total_sell = 0
    info = []
    def __init__(self, name):
        self.name = name
        self.outletsell = 0
        SultansDine.total += 1
    def sellQuantity(self, quantity):
        if quantity < 10:
            self.outletsell = quantity*300
        elif quantity < 20:
            self.outletsell = quantity*350
        else:
            self.outletsell = quantity*400
        
        SultansDine.total_sell += self.outletsell
        
    def branchInformation(self):
        print(f"Branch Name: {self.name}")
        print(f"Branch Sell: {self.outletsell} Taka")
        SultansDine.info.append([self.name, self.outletsell])
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {SultansDine.total}")
        print(f"Total Sell: {SultansDine.total_sell} Taka")
        for item in SultansDine.info:
            print(f"Branch Name: {item[0]}, Branch Sell: {item[1]} Taka ")
            print(f"Branch consists of total sell's: {((item[1]/SultansDine.total_sell)*100):.2f}%")


    
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
