```python
class RickMote:
    def __init__(self):
        self.sta = False
        self.channel_list = [0, 2, 3, 6, 7, 9]
        self.current_channel = 0
        self.channel_index = 0
        self.volume = 3
    def power(self):
        if self.sta == False:
            self.sta = True
        else:
            self.sta = False
    def changeChannel(self, signal = 1):
        if self.sta == True:
            if signal == 1:
                if self.channel_index + 1 == len(self.channel_list):
                    self.channel_index = 0 
                else:
                    self.channel_index + 1
                self.current_channel = self.channel_list[self.channel_index]
            elif signal in self.channel_list:
                self.current_channel = signal
                self.channel_index = self.channel_list.index(signal)
        else:
            print("Power is turned off. Cannot change channel.")
    def changeVolumeLevel(self, signal = 1):
        if self.sta == True:
            if self.volume + signal > 0 :
                self.volume = self.volume + signal
            else:
                self.volume = 0
        else:
            print("Power is turned off. Cannot change volume.")
    def showInfo(self):
        print("ID Cable Box Status:")
        if self.sta == True:
            print(f"Cable Box is: {self.sta}")
            print(f"Channet: {self.current_channel}")
            print(f"Volume: {self.volume}")
        else:
            print(f"Cable Box is: {self.sta}")

oTV = RickMote()
oTV.power()
print("1.#######################")
oTV.showInfo()
print("2.#######################")
oTV.changeChannel()
oTV.changeVolumeLevel()
oTV.showInfo()
print("3.#######################")
oTV.power()
oTV.showInfo()
print("4.#######################")
oTV.power()
oTV.changeChannel(4)
oTV.changeVolumeLevel(3)
oTV.showInfo()
print("5.#######################")
oTV.changeVolumeLevel(-2)
oTV.showInfo()
print("6.#######################")
oTV.power()
oTV.changeChannel(9)
oTV.changeVolumeLevel(-1)
oTV.showInfo()
print("7.#######################")
oTV.power()
oTV.changeChannel(11)
oTV.showInfo()
```

```python
#8
class box:
    def __init__(self, index):
        print("Creating a Box!")
        self.height = index[0]
        self.width = index[1]
        self.breadth = index[2]
        self.cube = self.height * self.width * self.breadth
        print(f"Volume of the box is {self.cube} cubic units.")

print("Box 1")
b1 = box([10,10,10])
print("=========================")
print("Height:", b1.height)
print("Width:", b1.width)
print("Breadth:", b1.breadth)
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
print("-------------------------")
print("Box 3")
b3 = b2
print("Height:", b3.height)
print("Width:", b3.width)
print("Breadth:", b3.breadth)
```

```python
#9
class buttons:
    def __init__(self, word, spaces, border):
        self.word = word
        self.spaces = spaces
        self.border = border
        self.top_bottom = 1 + self.spaces + len(self.word) + self.spaces + 1
        print(f"{self.word} Button Specifications:")
        print("Button name:", self.word)
        print("Number of the border characters for the top and the bottom:", self.top_bottom)
        print("Number of spaces between the left side border and the first character of the button name:", self.spaces)
        print("Number of spaces between the right side border and the last character of the button name:", self.spaces)
        print("Characters representing the borders:", self.border)
        n = " "
        print(f"{self.top_bottom*self.border}\n{self.border}{self.spaces*n}{self.word}{self.spaces*n}{self.border}\n{self.top_bottom*self.border}")

word = "CANCEL"  
spaces = 10
border = 'x'
self = buttons(word, spaces, border)
print("=======================================================")
b2 = buttons("Notify",3, '!')
print("=======================================================")
b3 = buttons('SAVE PROGRESS', 5, '$')
```
