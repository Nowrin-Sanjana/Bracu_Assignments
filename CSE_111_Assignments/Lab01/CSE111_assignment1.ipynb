```python
#Task 1

height = int(input())
weight = int(input())
def AmiKenoMota(height, weight):
    height = (height/100)
    bmi = weight/height**2

    if bmi < 18.5:
        state = "Underweight"
    elif 18.5 <= bmi <= 24.9:
        state = "Normal"
    elif 25 <= bmi <= 30:
        state = "Overweight"
    elif bmi > 30:
        state = "Obese"
    return f"Score is {bmi:.1f}. You are {state}"
print(AmiKenoMota(height, weight))
```

```python
#Task 4

def replace_domain(email, new_domain, old_domain= "kaaj.com"):
    lst = email.split("@")
    first_part  = lst[0]
    new_email = first_part + "@" + new_domain
    if lst[1] == old_domain:
        return f"Changed: {new_email}"
    else:
        return f"Uhanged: {new_email}"


print(replace_domain('alice@kaaj.com','sheba.xyz','kaaj.com'))
print(replace_domain('bob@sheba.xyz','sheba.xyz'))
```

```python
#Task 5


def palindrome_check(string):
    string1 = ""
    reverse = ""
    for char in string:
        if char != " ":
            string1 += char

    for item in range(len(string1)-1, -1,-1):
        reverse += string1[item]
    
    if string1 == reverse:
        return "Palindrome"
    else:
        return "Not a palindrome"
print(palindrome_check("madam"))
```

```python
#Task 3
def UB_Jupmer(List):
    ref_list = [num for num in range(1, len(List))]
    abs_diff = []

    for i in range(len(List) - 1):
        elem = List[i] - List[i + 1]
        elem = abs(elem)
        abs_diff.append(elem)

    if sorted(ref_list) == sorted(abs_diff):
        return True
    else:
        return False


user_input = input()
result = []
while user_input != "STOP":
    List = [int(num) for num in user_input.split()]

    if UB_Jupmer(List):
        result.append("UB Jumper")
    else:
        result.append("Not UB Jumper")

    user_input = input()

for output in result:
    print(output)
```

```python
#Task 1
str_list1 = input().split(", ")
str_list2 = input().split(", ")
dict1 = {}
dict2 = {}
for item in str_list1:
    key,value = item.split(": ")
    dict1[key] = int(value)
for item in str_list2:
    key,value = item.split(": ")
    dict2[key] = int(value)

new = {}
for key, value in dict1.items():
    if key in dict2:
        new[key] = value + dict2[key]
    else:
        new[key] = value
for key in dict2:
    if key not in new:
        new[key] = dict2[key]
print(new)
value = tuple(sorted(new.values()))
new_tup = []
for item in value:
    if item not in new_tup:
        new_tup.append(item)
print(tuple(new_tup))
```
