```python
#1
user = input()
upper = 0
lower = 0

for item in user:
    if 65<= ord(item)<= 90:
        upper += 1
    else:
        lower += 1
new = ""
if lower >= upper:
    for item in user:
        if 97<= ord(item)<= 122:
            new += item
        elif 65<= ord(item)<= 90:
            new += chr(ord(item)+32)  
            
else:
    for item in user:
        if 65<= ord(item)<= 90:
            new += item
        elif 97<= ord(item)<= 122:
            new += chr(ord(item)-32)      
print(new)
```

```python
#4
user = input()
user1 = user[:user.find(",")]
user2 = user[user.find(" ")+1:]
new = ""
for item in user1:
    if item in user2:
        new += item
for item in user2:
    if item in user1:
        new += item
if new != "":
    print(new)
else:
    print("Nothing in common.")
```

```python
# 2
user = int(input())
max = None
max_sum_list = []
for item in range(user):
    string_int = input()
    split_string = string_int.split(" ")
    new = []
    for char in split_string:
        new.append(int(char))
    max_sum = 0
    for item in new:
        max_sum += item
    if max == None or max_sum > max:
        max = max_sum
        max_sum_list = new
print(max)
print(max_sum_list)
```

```python
# 1
s_dict1 = input().split(", ")
s_dict2 = input().split(", ")
dict1 = {}
dict2 = {}
for item in s_dict1:
    key,value = item.split(": ")
    dict1[key] = int(value)
for item in s_dict2:
    key,value = item.split(": ")
    dict2[key] = int(value)
new = {}
for key in dict1:
    if key in dict2:
        new[key] = dict1[key] + dict2[key]
    else:
        new[key] = dict1[key]
for key in dict2:
    if key not in new:
        new[key] = dict2[key]
print(new)
new_tup = []
value = tuple(sorted(new.values()))
for item in value:
    if item not in new_tup:
        new_tup.append(item)
print(tuple(new_tup))
```

```python
# 3
def foodpanda(burger, loc = "Mohakhali"):
    if burger == "BBQ Chicken Cheese Burger":
        meal = 250
    elif burger == "Beef Burger":
        meal = 170
    elif burger == "Naga Drums":
        meal = 200
    if loc != "Mohakhali":
        delivery_c = 60
    else:
        delivery_c = 40
    tax = meal * 8/100

    total = meal + delivery_c + tax
    return total
string = input().split(",")
if len(string) == 2:
    print(foodpanda(string[0], string[1]))
else:
    print(foodpanda(string[0]))
```

```python
class Calculator:
    def __init__(self, first_value, operator, second_value):
        print("Let's Calculate!")
        self.first_value = first_value
        self.operator = operator
        self.second_value = second_value

        print(f"Value 1: {self.first_value}")
        print(f"Operator: {self.operator}")
        print(f"Value 2: {self.second_value}")

        if self.operator == "+":
            Calculator.add(self)
        elif self.operator == "-":
            Calculator.subtract(self)
        elif self.operator == "*":
            Calculator.multiply(self)
        elif self.operator == "/":
            Calculator.divide(self)

    def add(self):
        print(f"Result : {self.first_value + self.second_value}")

    def subtract(self):
        print(f"Result : {self.first_value - self.second_value}")

    def multiply(self):
        print(f"Result : {self.first_value * self.second_value}")

    def divide(self):
        print(f"Result : {self.first_value / self.second_value}")


first_value = int(input())
operator = input()
second_value = int(input())

calc = Calculator(first_value, operator, second_value)
```
