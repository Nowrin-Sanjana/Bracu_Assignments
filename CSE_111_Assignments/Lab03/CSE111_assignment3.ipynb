```python
#Task 3
class Calculator:
    def __init__(self):
        print("Calculator is ready!")
    def calculate(self, value1, value2, symbol):
        self.value1 = value1
        self.value2 = value2
        self.symbol = symbol
        if self.symbol == "+":
            output = self.value1 + self.value2
        elif self.symbol == "-":
            output = self.value1 - self.value2
        elif self.symbol == "*":
            output = self.value1*self.value2
        elif self.symbol == "/":
            output = self.value1/self.value2
        self.output = output
        return self.output
    def showCalculation(self):
        print(f"{self.value1} {self.symbol} {self.value2} = {self.output}")

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()

```

```python
#Task 13

class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}
    def calculateGPA(self, listcourse, intake):
        totalgrade = 0
        course = []
        for item in listcourse:
            subname = item[:item.find(":")]
            course.append(subname)
            grade = float(item[item.find(" ")+1:])
            totalgrade += grade
        gpa = round(((totalgrade*3.0)/(len(listcourse)*3.0)),2)
        self.grades[intake] = {tuple(course):gpa}
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        for key in self.grades:
            print(f"Courses taken in {key}:")
            for subject, point in self.grades[key].items():
                for item in subject:
                    print(item)
                print(f"GPA: {point}")

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
