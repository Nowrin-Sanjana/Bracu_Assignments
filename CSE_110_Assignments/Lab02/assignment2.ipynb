```python
# Task 28
lower = int(input("Enter the starting digit: "))
upper = int(input("Enter the end digit: "))
count = 0
count1 = 0
prime = ""
perfect = ""
print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            count += 1
            prime += str(num)+ ","
    total = 0
    if num > 0:
        for c in range(1, num):
            if num % c == 0:
                total += c
        if total == num:
            count1+= 1
            perfect += str(num)
print("Found", count, "prime number")
print("Found", count1, "perfect number")
print("Prime numbers:", prime, end = " ")
print()
print("Perfect numbers:", perfect, end = " ")          
```
