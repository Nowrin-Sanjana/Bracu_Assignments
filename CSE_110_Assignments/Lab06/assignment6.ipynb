```python
def count(string):
    unique=[]
    result={}
    length=len(string)
    for x in string:
        if x not in unique:
            unique.append(x)
    for val in unique:
        result[val]=[]
        for x in range(length):
            if string[x]==val:
                result[val].append(x)
                result[val].append(x-length)
    return result

print(count("pythonbook"))
```

```python
def occurence_counter(sentence):          
    """Takes in a string and produces a dictionary containing the number of occurence of each word in a string."""              
    l = sentence.split()     
    word_dict  = {}          
    for word in l:         
        if word in word_dict:             
            word_dict[word] = word_dict[word]  + 1;         
        else:             
            word_dict[word] = 1;          
    sorted_dict = sorted(word_dict.items(), key = lambda kv:(kv[1], kv[0]))     
    print(sorted_dict)           
text = input("Enter a space separated string:- ")  
occurence_counter(text)
 
```
