```python
giv = input().strip('"').split(", ")
magic = []
not_magic = []
final = []
for item in giv:
    even = 0
    odd = 0
    for idx in range(len(item)):
        if idx%2 == 0:
            even += int(item[idx])
        else:
            odd += int(item[idx])
    if even == odd:
        magic.append(int(item))
    else:
        not_magic.append(int(item))
final.append(tuple(magic))
final.append(tuple(not_magic))
final = tuple(final)
print(final)
```

```python
dict1 = {"a":[5,2,55,17],"P":(11,121,222),"B":(37,53,71),"c":[45,92,50]}
new = {}
for key,val in dict1.items():
    for item in val:
        c = 0
        for idx in range(2,item):
            if item%idx == 0:
                c = 1
                break
        if c == 0:
            if key not in new:
                new[key] = [item]
            else:
                new[key].append(item)
        if key not in new:
                new[key] = []
                
for k in new:
    if k == k.upper():
        new[k] = tuple(new[k])
print(f"dict_primes = {new}")
```
