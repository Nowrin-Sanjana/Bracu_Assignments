```python
#1

n = input()
alphabet = 0
digit = 0
char = 0
for item in n:
    if 97<= ord(item) <=122 or 65<= ord(item) <=90:
        alphabet += 1
    elif 48<= ord(item) <=57:
        digit += 1
    else:
        char += 1
print("The number of Alphabets in the string is: ", alphabet)
print("The number of Digits in the string is : ", digit)
print("The number of Special characters in the string is: ", char)
```

```python
#2

n = input()

word = n.split(" ")
largest = smallest = word[0]
for item in range(len(word)):
    if len(largest) < len(word[item]):
        largest = word[item]
    if len(smallest) > len(word[item]):
        smallest = word[item]
print("The largest word is '",largest, "' and the smallest word is '", smallest, "'")
```
