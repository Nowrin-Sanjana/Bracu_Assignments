```python
list1 = [int(item) for item in input()[1:-1].split(", ")]
list2 = [int(item) for item in input()[1:-1].split(", ")]
sort = []
new1 = sorted(list1)
new2 = sorted(list2)
for item in new1:
    sort.append(item)
for item in new2:
    sort.append(item)   
new = sorted(sort)
print(f"Sorted list = {new}")
if len(new)%2 != 0:
    number = int((len(new)+1)/2)
    print(f"Median = {new[number-1]}")
else:
    index1 = int(len(new)//2)-1
    index2= int(((len(new))/2)+1)
    median=float((new[index1]+new[index2])/2)
    print(f"Median = {median}")
```

```python
list_one = [-10, 15, 2, 4, -4, 7, -8]
pos = []
neg = []
for item in list_one:
    if item > 0:
        pos.append(item)
    else:
        neg.append(item)
sn1 = 0
sn2 = 0
div = None
for j in range(len(neg)):
    min = pos[0] + neg[j]
    if div == None or div > min:
        div = min
    for k in range(len(pos)):
        min = pos[k] + neg[j]
        min = abs(min)
        if min<div:
            sn1 = pos[k]
            sn2 = neg[j]
            div = min
print(sn1, sn2)          
```
