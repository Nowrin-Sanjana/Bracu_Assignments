```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/input03.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/output03.txt","w")
length = int(inp.readline())
start_end = []
for i in range(length):
    lines = list(map(int, inp.readline().split(" ")))
    start_end.append(lines)
def insertion_sort(s_e):
    for i in range(1, length):
        key_s_e = s_e[i]
        j = i - 1
        while j >= 0 and key_s_e[1] < s_e[j][1]:
            s_e[j + 1] = s_e[j]
            j -= 1
        s_e[j + 1] = key_s_e
    return s_e
sorted_s_e = insertion_sort(start_end)
count=1
index=0
task = [sorted_s_e[0]]
for i in range(1,length):
    if sorted_s_e[index][1]<=sorted_s_e[i][0]:
        count+=1
        index=i
        task.append(sorted_s_e[i])
output.write(f"{count}\n")
for item in task:
    output.write(f"{item[0]} {item[1]}\n")
inp.close()
output.close()
```
