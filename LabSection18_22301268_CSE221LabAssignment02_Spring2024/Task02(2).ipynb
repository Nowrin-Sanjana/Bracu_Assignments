```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/input02(2).txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/output02(2).txt","w")
var = inp.read().split("\n")
len1, len2 = int(var[0]), int(var[2])
l1 = list(map(int, var[1].split(" ")))
l2 = list(map(int, var[3].split(" ")))
def merge(a, b):
        i = 0
        j = 0
        l = []
        merged_arr = []
        while i < len1 and j < len2:
            if a[i] <= b[j]:
                l.append(a[i])
                i += 1
            else:
                l.append(b[j])
                j += 1
        if i>len1-1:
            merged_arr = l + b[j:]
        if j>len2-1:
            merged_arr = l + a[i:]
        return merged_arr
sorted_arr = merge(l1, l2)
for item in sorted_arr:
    output.write("".join(str(item)+" "))
inp.close()
output.close()
```
