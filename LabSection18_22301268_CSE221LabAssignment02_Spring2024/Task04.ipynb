```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/input04.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/output04.txt","w")
line = list(map(int,inp.readline().split(" ")))
length, people = line[0], [[0]*2]*line[1]
start_end = []
for i in range(length):
    lines = list(map(int, inp.readline().split(" ")))
    start_end.append(lines)
print(start_end)
def merge_sort(arr):
    if len(arr)==1:
        return arr
    else:
        mid=len(arr)//2
        a = merge_sort(arr[:mid])
        b = merge_sort(arr[mid:])
        return merge(a,b)
def merge(a, b):
    i = 0
    j = 0
    l = []
    merged_arr = []
    while i < len(a) and j < len(b):
        if a[i][1] <= b[j][1]:
            l.append(a[i])
            i += 1
        else:
            l.append(b[j])
            j += 1
    if i>len(a)-1:
        merged_arr = l + b[j:]
    if j>len(b)-1:
        merged_arr = l + a[i:]
    return merged_arr
sorted_arr = merge_sort(start_end)
count = 0
for activity in sorted_arr:
    selected_person = None
    min_difference = float('inf')
    for j in range(len(people)):
        if people[j][1] <= activity[0]:
            difference = activity[0] - people[j][1]
            if difference < min_difference:
                min_difference = difference
                selected_person = j
    if selected_person != None:
        people[selected_person] = activity
        count += 1
output.write(f"{count}")
inp.close()
output.close()
```
