```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/input01(2).txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/output01(2).txt","w")
variable = inp.read().split("\n")[0:]
length,target = list(map(int,variable[0].split()))
digits = list(map(int, variable[1].split()))
flag = True
start = 0
end = length-1
while start<end:
    if digits[start] + digits[end] == target:
        output.write(f"{start+1} {end + 1}")
        flag = False
        break
    elif digits[start] + digits[end] < target:
        start+=1
    elif digits[start] + digits[end] > target:
        end-=1
if flag==True:
    output.write("IMPOSSIBLE")
inp.close()
output.close()
```
