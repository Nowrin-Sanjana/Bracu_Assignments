```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment05_Spring2024/input1(b)_1.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment05_Spring2024/output1(b)_1.txt", "w")

vertices, edges = map(int, inp.readline().split())
var = inp.read().split('\n')
source = int(var[0][0])

queue = []
nodes = []
in_degree = [0] * (vertices + 1)
G = {}

for i in range(1, vertices + 1):
    G[i] = []

for j in range(edges):
    u, v = list(map(int, var[j].split()))
    G[u].append(v)
    in_degree[v] += 1
print(G)
in_degree[0] = 9999
print(in_degree)
def BFS(G, vertices):
    for i in range(1, vertices + 1):
        if in_degree[i] == 0:
            queue.append(i)
    print(queue)
    while queue:
        current = queue.pop(0)
        print(current)
        nodes.append(current)
        for neighbour in G[current]:
            print(neighbour)
            in_degree[neighbour] -= 1
            print(in_degree)
            if in_degree[neighbour] == 0:
                queue.append(neighbour)
                print(queue)

BFS(G, vertices)

if len(nodes) != vertices:
    output.write("IMPOSSIBLE")
else:
    for elem2 in nodes:
        output.write(str(elem2) + " ")
print(in_degree)
print(nodes)
inp.close()
output.close()
```
