```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment05_Spring2024/input1(a)_2.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment05_Spring2024/output1(a)_2.txt", "w")

vertices, edges = map(int, inp.readline().split(" "))
var = inp.read().split('\n')
print(var)
source = int(var[0][0])
stack = []
visited = [0] * (vertices + 1)
G = {}

for i in range(1, vertices + 1):
    G[i] = []

for j in range(edges):
    u, v = list(map(int, var[j].split(" ")))
    G[u].append(v)
print(G)
flag = True
def DFS(G, u):
    global flag
    visited[u] = 1
    stack.append(u)
    for neighbour in G[u]:
        if visited[neighbour] == 0:
            DFS(G, neighbour)
        elif visited[neighbour] == 1:
            flag = False
            return
    visited[u] = 2
DFS(G, source)

for k in range(1, vertices + 1):
    if visited[k] == 0:
        DFS(G, k)
print(visited)
if flag:
    for elem2 in stack:
        output.write(str(elem2) + " ")
else:
    output.write("IMPOSSIBLE")
inp.close()
output.close()
```
