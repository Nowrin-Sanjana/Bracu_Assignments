```python
with open("/content/drive/MyDrive/Semester/6th Semester/CSE221/18_22301268_CSE221_Lab01_Spring2024/input1b.txt", "r") as inp:
    output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/18_22301268_CSE221_Lab01_Spring2024/output1b.txt", "w")
    inp_num = int(inp.readline())
    for i in range(inp_num):
        result = 0
        l = inp.readline().split()
        if l[2]=="+":
            result=int(l[1])+int(l[3])
            output.write(f"The result of {l[1]} + {l[3]} is {result}\n")
        elif l[2]=="-":
            result=int(l[1])-int(l[3])
            output.write(f"The result of {l[1]} - {l[3]} is {result}\n")
        elif l[2]=="/":
            result=int(l[1])/int(l[3])
            output.write(f"The result of {l[1]} / {l[3]} is {result}\n")
        elif l[2]=="*":
            result=int(l[1])*int(l[3])
            output.write(f"The result of {l[1]} * {l[3]} is {result}\n")
    output.close()
```
