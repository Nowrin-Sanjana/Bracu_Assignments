```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment06_Spring2024/input1_2.txt","r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment06_Spring2024/output1_2.txt","w")
vertices, edges = map(int, inp.readline().split())
var = inp.read().split('\n')
source = int(var[-1])
G = {}
for i in range(1, vertices+1):
    weight = [float('inf')] * (i+1)
    G[i] = []

weight[source] = 0
for j in range(edges):
    u, v, w = list(map(int, var[j].split(" ")))
    G[u].append([v, w])
print(G)
print(weight)
def shortest_path_Dijkstra(G, s):
    visited = [None] * (vertices+1)
    queue = []
    queue.append(s)
    visited[s] = 1
    while queue:
        current = queue.pop(0)
        for neighbor in G[current]:
            d_check = weight[current]+ neighbor[1]
            print(d_check)
            if d_check < weight[neighbor[0]]:
                weight[neighbor[0]] = d_check
                print(weight)
            if visited[neighbor[0]]!= 1:
                queue.append(neighbor[0])
                visited[neighbor[0]] = 1

shortest_path_Dijkstra(G, source)

for k in range(1,len(weight)):
    if weight[k] == float('inf'):
        output.write("-1")
    else:
        output.write(str(weight[k])+" ")
inp.close()
output.close()
```
