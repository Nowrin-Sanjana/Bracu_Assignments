```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment06_Spring2024/input3_2.txt","r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment06_Spring2024/output3_2.txt","w")
vertices, edges = map(int, inp.readline().split(" "))
var = inp.read().split('\n')
source = 1
G = {}
for i in range(1, vertices+1):
    weight = [float('inf')] * (i+1)
    G[i] = []
weight[source] = 0
for j in range(edges):
    u, v, w = list(map(int, var[j].split(" ")))
    G[u].append([v, w])
maximum = 0
print(G)
def shortest_path_Dijkstra(G, s):
    visited = [None] * (vertices+1)
    queue = []
    queue.append(s)
    visited[s] = 1
    while queue:
        current = queue.pop(0)
        for neighbor in G[current]:
            maximum = max(weight[current], neighbor[1])
            if maximum < weight[neighbor[0]]:
                weight[neighbor[0]] = maximum
            if visited[neighbor[0]]!= 1:
                queue.append(neighbor[0])
                visited[neighbor[0]] = 1
shortest_path_Dijkstra(G, source)
if weight[vertices] == float('inf'):
    output.write("Impossible")
else:
    output.write(f"{weight[vertices]}")
print(weight)
inp.close()
output.close()
```
