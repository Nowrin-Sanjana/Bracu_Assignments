```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment06_Spring2024/input2_3.txt","r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment06_Spring2024/output2_3.txt","w")
vertices, edges = map(int, inp.readline().split())
var = inp.read().split('\n')
s_a, s_b = map(int, var[-1].split(" "))
G = {}

for i in range(1, vertices+1):
    weight_a = [float('inf')] * (i+1)
    weight_b = [float('inf')] * (i+1)
    G[i] = []
weight_a[s_a] = 0
weight_b[s_b] = 0
for j in range(edges):
    u, v, w = list(map(int, var[j].split(" ")))
    G[u].append([v, w])
def Metting_point_time_a_b(G, s, weight):
    visited = [None] * (vertices+1)
    queue = []
    queue.append(s)
    visited[s] = 1
    while queue:
        current = queue.pop(0)
        for neighbor in G[current]:
            d_check = weight[current]+ neighbor[1]
            if d_check < weight[neighbor[0]]:
                weight[neighbor[0]] = d_check
            if visited[neighbor[0]]!= 1:
                queue.append(neighbor[0])
                visited[neighbor[0]] = 1
    return(weight)
path_a = Metting_point_time_a_b(G, s_a, weight_a)
path_b = Metting_point_time_a_b(G, s_b, weight_b)
print(path_a)
print(path_b)
meeting_point=[]
for idx in range(len(path_a)):
    if path_a[idx]!=float("inf") and path_b[idx]!=float("inf") and path_a[idx]!=0 and path_b[idx]!=0:
        if path_a[idx]>=path_b[idx]:
          meeting_point.append([path_a[idx], idx])
        else:
          meeting_point.append([path_b[idx], idx])
min = float("inf")
if  meeting_point!= []:
    for item in meeting_point:
        if item[0]<min:
            min=item[0]
            m_time = item[1]
            # print(m_time)
    output.write(f"Time {min}\nNode {m_time}")
else:
    output.write("Impossible")
inp.close()
output.close()
```
