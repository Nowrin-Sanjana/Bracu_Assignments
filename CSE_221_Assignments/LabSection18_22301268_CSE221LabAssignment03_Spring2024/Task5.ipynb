```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/input5.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/output5.txt", "w")
length = int(inp.readline())
unsorted_list = list(map(int, inp.readline().split(" ")))
def partition(A, pivot, right):
    x = A[right]
    i = pivot - 1
    for j in range(pivot, right):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[right] = A[right], A[i + 1]
    return i + 1
def quick_sort(A, pivot, right):
    if pivot < right:
        q = partition(A, pivot, right)
        quick_sort(A, pivot, q - 1)
        quick_sort(A, q + 1, right)
quick_sort(unsorted_list, 0, length - 1)
for item in unsorted_list:
    output.write("".join(str(item))+" ")
inp.close()
output.close()
```
