```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/input2.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/output2.txt", "w")
length = int(inp.readline())
unsorted_list = list(map(int, inp.readline().split(" ")))
def merge_sort(l):
    if len(l) == 1:
        return l
    else:
        mid = len(l) // 2
        a = merge_sort(l[:mid])
        b = merge_sort(l[mid:])
        return merge(a, b)
max_val = 0
def merge(a, b):
    if a <= b:
        max_val = b
    else:
        max_val = a
    return max_val
maximum_num = merge_sort(unsorted_list)
output.write(f"{maximum_num[0]}")

inp.close()
output.close()
```
