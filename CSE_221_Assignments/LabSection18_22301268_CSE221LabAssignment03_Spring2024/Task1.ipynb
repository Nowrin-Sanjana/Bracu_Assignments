```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/input1.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/output1.txt", "w")
length = int(inp.readline())
unsorted_list = list(map(int, inp.readline().split(" ")))
def merge_sort(l):
    if len(l) == 1:
        return l
    else:
        mid = len(l) // 2
        a = merge_sort(l[:mid])
        b = merge_sort(l[mid:])
        return merge(a, b)
def merge(a, b):
    i = 0
    j = 0
    l = []
    merged_list = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            l.append(a[i])
            i += 1
        else:
            l.append(b[j])
            j += 1
    if i>len(a)-1:
        merged_arr = l + b[j:]
    if j>len(b)-1:
        merged_arr = l + a[i:]
    return merged_arr

sorted_list = merge_sort(unsorted_list)
for elem in sorted_list:
    output.write(f"{elem} ")
inp.close()
output.close()
```
