```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/input01(1).txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment02_Spring2024/output01(1).txt","w")
variable = inp.read().split("\n")[0:]
length,target = list(map(int,variable[0].split()))
digits = list(map(int, variable[1].split()))
flag = True
for i in range(0, length):
    for j in range(i+1, length):
        if digits[i]+digits[j] == target:
            output.write(f"{i+1} {j+1}")
            flag=False
            break
    if flag==False:
        break
if flag==True:
    output.write("IMPOSSIBLE")
inp.close()
output.close()
```
