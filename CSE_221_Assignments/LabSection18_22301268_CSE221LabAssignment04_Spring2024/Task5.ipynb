```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/input5_2.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/output5_2.txt", "w")
from collections import deque
vertices, edges, destination = map(int, inp.readline().split(" "))
var=inp.read().split('\n')
G = {}
for k in range(1, vertices+1):
    G[k]=[]
for iter in range(edges):
    u,v = list(map(int,var[iter].split(" ")))
    G[u].append((v))
    G[v].append((u))

def shortest_path(G, s, d):
    if s == d:
        return [s], 0
    queue = [(s, [s], 0)]
    visited = []
    while queue:
        current, path, time = queue.pop(0)
        if current == d:
            return path, time
        visited.append(current)
        for neighbor in G[current]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor], time + 1))
                visited.append(neighbor)
path,time = shortest_path(G, 1, destination)
output.write(f"Time: {time}\n")
output.write(f"Shortest Path:")
for item in path:
    output.write(f" {item}")
inp.close()
output.close()
```
