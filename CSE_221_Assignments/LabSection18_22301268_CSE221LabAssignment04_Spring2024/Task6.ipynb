```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/input6_7.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/output6_7.txt", "w")
vertices, edges = map(int, inp.readline().split(" "))

G = {}
for k in range(1, vertices+1):
    G[k]=[]
grid=[list(inp.readline().strip())]
for i in range(vertices-1):
    grid.append(list(inp.readline().strip()))

print(grid)
def bfs(grid, x, y, block="#"):
    row_num = len(grid)
    col_num = len(grid[0])
    count = 0
    if grid[x][y] == block:
        return count
    queue = []
    queue.append((x,y))
    while len(queue)> 0:
        x , y = queue.pop(0)
        if x<0 or x >= row_num or y<0 or y >= col_num or grid[x][y] == block:
            continue
        else:
            if grid[x][y] == "D":
                count += 1
            grid[x][y] = block
            queue.append((x+1, y))
            queue.append((x-1, y))
            queue.append((x, y+1))
            queue.append((x, y-1))
    return count
def flood_fill(grid):
    max_l = [0]
    row = len(grid)
    col = len(grid[0])
    for i in range(row):
        for j in range(col):
            total = 0
            if grid[i][j] == ".":
                total += bfs(grid, i, j)
                max_l.append(total)
    return max(max_l)
max_d = flood_fill(grid)
output.write(str(max_d))
inp.close()
output.close()
```
