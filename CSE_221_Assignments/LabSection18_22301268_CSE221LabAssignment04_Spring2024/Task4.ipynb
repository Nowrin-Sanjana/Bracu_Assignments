```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/input4_5.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/output4_5.txt", "w")
vertices, edges = map(int, inp.readline().split(" "))
var=inp.read().split('\n')
source = int(var[0][0])
G = {}
for k in range(1, vertices+1):
    G[k]=[]
for iter in range(edges):
    u,v = list(map(int,var[iter].split(" ")))
    G[u].append((v))
# print(G)

def BFS(G, s):
    visited = [0] * (vertices+1)
    queue = []
    queue.append(s)
    visited[s] = 1
    flag = True
    while queue:
        current = queue.pop(0)
        for neighbor in G[current]:
            if visited[neighbor] == 0:
                queue.append(neighbor)
                visited[neighbor] = 1
                flag = False
            if visited[neighbor]>0:
                for elem in G[neighbor]:
                    if visited[elem] == 1:
                        for item in G[elem]:
                            if visited[item]==1:
                                flag = True
                                break
    if flag == True:
        output.write("YES")
    else:
        output.write("NO")
BFS(G, source)
inp.close()
output.close()
```
