```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/input3_4.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment04_Spring2024/output3_4.txt", "w")
vertices, edges = map(int, inp.readline().split(" "))
var=inp.read().split('\n')
source = int(var[0][0])
visited = [None] * (vertices+1)
def DFS(G, u):
    visited[u] = 1
    output.write(str(u) + " ")
    for v in G[u]:
        if visited[v] != 1:
            DFS(G, v)
G = {}
for k in range(1, vertices+1):
    G[k]=[]
    if visited[k]!=1:
        DFS(G, k)
for iter in range(edges):
    u,v = list(map(int,var[iter].split(" ")))
    G[u].append((v))
    G[v].append((u))
DFS(G, source)
inp.close()
output.close()
```
