```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/input1_2.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/output1_2.txt", "w")
p, q = list(map(int, inp.readline().split(" ")))
var = inp.read().split("\n")
fr_size = [1]*(p+1)
parent = list(range(p+1))
res = []
def find(parent, i):
    if parent[i] == i:
        return i
    else:
        parent[i] = find(parent, parent[i])
        return parent[i]
def union(parent, fr_size, u, v):
    u_rep = find(parent, u)
    v_rep = find(parent, v)
    if u_rep != v_rep:
        if fr_size[u_rep] < fr_size[v_rep]:
            parent[u_rep] = v_rep
            fr_size[v_rep] += fr_size[u_rep]
            return fr_size[v_rep]
        else:
            parent[v_rep] = u_rep
            fr_size[u_rep] += fr_size[v_rep]
            return fr_size[u_rep]
    return fr_size[u_rep]
for i in range(q):
    u, v = list(map(int, var[i].split(" ")))
    c_size = union(parent, fr_size, u, v)
    res.append(c_size)
for k in res:
    output.write(f"{k}\n")
inp.close()
output.close()
```
