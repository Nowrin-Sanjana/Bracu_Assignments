```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/input3_1.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/output3_1.txt", "w")
n = int(inp.readline())
dw = [0] * (n + 1)
def climb_stairs(dw, n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    dw = [0] * (n + 1)
    dw[0] = 1
    dw[1] = 1
    for i in range(2, n + 1):
        dw[i] = dw[i - 1] + dw[i - 2]
    return dw[n]

res = climb_stairs(dw, n)
output.write(f'{res}')

inp.close()
output.close()
```
