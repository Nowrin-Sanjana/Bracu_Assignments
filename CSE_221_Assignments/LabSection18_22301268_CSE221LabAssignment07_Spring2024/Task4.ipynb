```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/input4_2.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/output4_2.txt", "w")

var = inp.readlines()

N, X = map(int, var[0].split(" "))
coins = list(map(int, var[1].split()))
dc = [float('inf')] * (X + 1)
dc[0] = 0
def minCoins(dc, coins, amount):
    for item in coins:
        for i in range(item, amount + 1):
            dc[i] = min(dc[i], dc[i - item] + 1)

    if dc[amount] != float('inf'):
      return dc[amount]
    else:
      return -1

res = minCoins(dc, coins, X)
output.write(f'{res}')

inp.close()
output.close()
```
