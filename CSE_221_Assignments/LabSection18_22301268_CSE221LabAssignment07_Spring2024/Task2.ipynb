```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/input2_2.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment07_Spring2024/output2_2.txt", "w")

from heapq import heappop, heappush

vertices, edges = list(map(int, inp.readline().split(" ")))
var = inp.read().split("\n")
G = {}
mst_G = {}
for i in range(vertices+1):
    G[i]=[]
    mst_G[i]=[]


#Creating x priority queue to pop the edges with the least weight first
priority_queue = []

for j in range(edges):
    u,v,w = list(map(int, var[j].split(" ")))
    G[u].append((v,w))
    G[v].append((u,w))
    heappush(priority_queue,(w,(u,v)))

parent = [-1]*(vertices+1)

def root(vertex):
    if parent[vertex]==-1:
        return vertex
    else:
        return root(parent[vertex])
c = 0
while priority_queue:
    w,edges = heappop(priority_queue)
    x, y = edges
    u = root(x)
    v = root(y)
    if u!=v:
        c += w
        parent[v] = u
        mst_G[x].append((y, w))

output.write(f"{c}")
inp.close()
output.close()
```
