```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/input6.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/output6.txt", "w")
length = int(inp.readline())
unsorted_list = list(map(int, inp.readline().split(" ")))
queries = int(inp.readline())
query_lines = list(map(int, inp.read().split("\n")))

def partition(A, pivot, right):
    x = A[right]
    i = pivot - 1
    for j in range(pivot, right):
        if A[j] <= x:
            i += 1
            A[i], A[j] = A[j], A[i]
    A[i + 1], A[right] = A[right], A[i + 1]
    return i + 1

def kth_smallest(A, low, high, k):
    if low <= high:
        pivot_idx = partition(A, low, high)
        if pivot_idx == k - 1:
            return A[pivot_idx]
        elif pivot_idx > k - 1:
            return kth_smallest(A, low, pivot_idx - 1, k)
        else:
            return kth_smallest(A, pivot_idx + 1, high, k)
for k in range(queries):
    query = query_lines[k]
    result = kth_smallest(unsorted_list , 0, length-1, query)
    output.write(f"{result}\n")
inp.close()
output.close()
```
