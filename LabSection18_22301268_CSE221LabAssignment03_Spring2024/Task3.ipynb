```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/input3.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/output3.txt", "w")
length = int(inp.readline())
unsorted_list = list(map(int, inp.readline().split(" ")))


def merge_sort(l, count=0):
    if len(l) == 1:
        return l, count
    else:
        mid = len(l) // 2
        a, c_a = merge_sort(l[:mid], count)
        b, c_b = merge_sort(l[mid:], count)
        merged_arr, merge_count = merge(a, b)
        c_total = c_a + c_b + merge_count
        print(c_a, c_b, merge_count)
        return merged_arr, c_total

def merge(a, b):
    i = 0
    j = 0
    merged_arr = []
    count = 0
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            merged_arr.append(a[i])
            i += 1
        else:
            merged_arr.append(b[j])
            j += 1
            count += len(a) - i

    if i > len(a) - 1:
        merged_arr.extend(b[j:])
    if j > len(b) - 1:
        merged_arr.extend(a[i:])

    return merged_arr, count


sorted_list, count = merge_sort(unsorted_list)
output.write(f"{count}")

inp.close()
output.close()
```
