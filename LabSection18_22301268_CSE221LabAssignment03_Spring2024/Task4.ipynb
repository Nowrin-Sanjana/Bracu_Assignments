```python
inp = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/input4.txt", "r")
output = open("/content/drive/MyDrive/Semester/6th Semester/CSE221/LabSection18_22301268_CSE221LabAssignment03_Spring2024/output4.txt", "w")
length = int(inp.readline())
unsorted_list = list(map(int, inp.readline().split(" ")))
def merge_sort(l):
    if len(l) == 1:
        return l
    else:
        mid = len(l) // 2
        a = merge_sort(l[:mid])
        b = merge_sort(l[mid:])
        return merge(a, b)
def merge(a, b):
    i = 0
    j = 0
    l = []
    merged_list = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            l.append(a[i])
            i += 1
        else:
            l.append(b[j])
            j += 1
    if i>len(a)-1:
        merged_arr = l + b[j:]
    if j>len(b)-1:
        merged_arr = l + a[i:]
    return merged_arr
def calc_max_sum(l):
    squared_val = []
    for index in range(length):
        val = l[index]
        squared_val.append((val ** 2, index))
    sorted_squared_val = merge_sort(squared_val)
    print(sorted_squared_val)
    i, j = 0, len(l) - 1
    max_addition = 0
    while i < j:
        if sorted_squared_val[j][1] <= i:
            j -= 1
        current_addition = l[i] + sorted_squared_val[j][0]
        if max_addition < current_addition:
            max_addition = current_addition
        i += 1
    return max_addition
maximum_sum = calc_max_sum(unsorted_list)
output.write(f"{maximum_sum}")
inp.close()
output.close()
```
