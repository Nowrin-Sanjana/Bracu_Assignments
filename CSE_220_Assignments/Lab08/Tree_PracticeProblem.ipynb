```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def are_identical_trees(root1, root2):
    if not root1 and not root2:
        return True
    if not root1 or not root2:
        return False
    return root1.val == root2.val and are_identical_trees(root1.left, root2.left) and are_identical_trees(root1.right, root2.right)

tree1 = TreeNode(1)
tree1.left = TreeNode(2)
tree1.right = TreeNode(3)

tree2 = TreeNode(1)
tree2.left = TreeNode(2)
tree2.right = TreeNode(3)

# Check if the trees are identical
if are_identical_trees(tree1, tree2):
    print("The trees are identical.")
else:
    print("The trees are not identical.")
```

```python
def printAncestors(root, key):
    if root is None:
        return False

    if root.elem == key:
        return True

    if printAncestors(root.left, key) or printAncestors(root.right, key):
        print(root.elem)
        return True

    return False

# Example usage:
# Create a binary tree
#        5
#       / \
#      2   9
#     / \   \

#    7   4   8
tree = BTNode(5)
tree.left = BTNode(2)
tree.right = BTNode(9)
tree.left.left = BTNode(7)
tree.left.right = BTNode(4)
tree.right.right = BTNode(8)

key = 8
print("Ancestors of", key, "are:")
printAncestors(tree, key)
```

```python
def skewed(root):
    if root is None:
        return True

    if root.left is not None and root.right is not None:
        return False

    if root.left is not None:
        return skewed(root.left)

    if root.right is not None:
        return skewed(root.right)

    return True

tree1 = BTNode(5)
tree1.left = BTNode(4)
tree1.left.right = BTNode(3)
tree1.left.right.left = BTNode(2)
print(skewed(tree1))

tree1 = BTNode(5)
tree1.left = BTNode(4)
tree1.left.right = BTNode(3)
tree1.left.right.left = BTNode(2)
tree1.left.right.right = BTNode(4)

print(skewed(tree1))  # Output: True (Skewed tree)

# Create another binary tree
#        5
#       / \
#      2   3
#     / \
#    1   8
tree2 = BTNode(5)
tree2.left = BTNode(2)
tree2.right = BTNode(3)
tree2.left.left = BTNode(1)
tree2.left.right = BTNode(8)
print(skewed(tree2))
```

```python
def minDepth(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    left_depth = minDepth(root.left)
    right_depth = minDepth(root.right)

    if root.left is None:
        return right_depth + 1
    elif root.right is None:
        return left_depth + 1

    return min(left_depth, right_depth) + 1

# Create a binary tree
#        3
#       / \
#      9  20
#         / \
#        15  7
tree = BTNode(3)
tree.left = BTNode(9)
tree.right = BTNode(20)
tree.right.left = BTNode(15)
tree.right.right = BTNode(7)
# tree.right.right.right = BTNode(7)

print("Minimum Depth of the binary tree:", minDepth(tree))
```
