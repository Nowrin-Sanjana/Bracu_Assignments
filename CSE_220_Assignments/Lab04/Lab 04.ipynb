```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group
        self.next = None
        self.prev = None
class WRM:
    def __init__(self):
        self.head = Patient(None, None, None, None)
        self.head.next = self.head
        self.head.prev = self.head

    def registerPatient(self, id, name, age, blood_group):
        registered = Patient(id, name, age, blood_group)
        patient_node = registered

        patient_node.next = self.head
        patient_node.prev = self.head.prev
        self.head.prev.next = patient_node
        self.head.prev = patient_node

        print(f"Patient {name} added successfully.")

    def ServePatient(self):
        if self.head.next == self.head:
            print("There is no patient to serve.")
        else:
            serving = self.head.next
            self.head.next = self.head.next.next
            self.head.next.prev = self.head
            print("Serving patient:", serving.name)

    def CancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head
        print("All appointment canceled.")

    def CanDoctorGoHome(self):
        if self.head.next == self.head:
            return True
        else:
            return False

    def ShowAllPatient(self):
        current_patient = self.head.next
        if self.head != current_patient:
            while current_patient != self.head:
                print("Patient Name:", current_patient.name, end = "\n")
                current_patient = current_patient.next
        else:
            print("No patient is in the waiting room.")

wrm = WRM()

while True:
    print()
    print("Welcome to Waiting Room Management System")
    print("Select your prefered service!!")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Check if Doctor can go home")
    print("5. Cancel All Appointments")
    print("6. Exit")
    print()
    preference = int(input("Enter your preference: "))

    if preference == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        print()
        wrm.registerPatient(id, name, age, blood_group)
    elif preference == 2:
        print()
        wrm.ServePatient()
    elif preference == 3:
        print()
        wrm.ShowAllPatient()
    elif preference == 4:
        print()
        if wrm.CanDoctorGoHome():
            print("Yes, doctor can go .")
        else:
            print("No, Doctor cannot go now.")
    elif preference == 5:
        print()
        wrm.CancelAll()
    elif preference == 6:
        break
    else:
        print("Invalid, please try again.")
```
