```python
def mean_value(arr):
  sum_of_number = 0
  sum_of_index = len(arr)
  mean = 0
  for item in range(0, len(arr)):
    sum_of_number+= arr[item]
  mean = sum_of_number/sum_of_index
  return mean
arr = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(f"{mean_value(arr):.12f}")
```

```python
import math
def standard_deviation(arr):
  sum_of_number = 0
  standard_deviation = 0
  for item in range(len(arr)):
    half_calc_of_formula = (arr[item] - mean_value(arr))**2
    sum_of_number += half_calc_of_formula
  standard_deviation = math.sqrt((1/(len(arr)-1)) * sum_of_number)
  return standard_deviation
arr = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(f"{standard_deviation(arr):.2f}")
```

```python
def new_array(arr):
  new = []
  times_standard_deviation = 1.5 * standard_deviation(arr)
  for item in range(len(arr)):
    new = [iter for iter in arr if abs(iter - mean_value(arr)) >times_standard_deviation]
    if len(new) > 0:
      return new
    else:
      return None

arr = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(new_array(arr))
```
